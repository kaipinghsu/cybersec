```python
import requests
import os
import subprocess
from urllib.parse import urlsplit

# 獲取文件名
def get_filename_from_url(url=None):
    if url is None:
        return None
    urlpath = urlsplit(url).path
    return os.path.basename(urlpath)

# 下載並加密文件
def download_and_zip_file(url, password):
    try:
        # 獲取文件名
        file_name = get_filename_from_url(url)
        if not file_name:
            raise ValueError("無法從URL獲取文件名")

        # 下載文件
        response = requests.get(url, verify=False)
        if response.status_code != 200:
            raise requests.exceptions.RequestException(f"下載失敗，狀態碼: {response.status_code}")

        # 保存文件
        with open(file_name, "wb") as f:
            f.write(response.content)

        # 壓縮並加密文件
        zip_file_name = file_name + ".zip"
        subprocess.run(['zip', '-P', password, zip_file_name, file_name], check=True)

        print(f"文件 '{file_name}' 下載並加密為 '{zip_file_name}'")

    except ValueError as ve:
        print(f"ValueError: {ve}")
    except requests.exceptions.RequestException as re:
        print(f"RequestException: {re}")
    except subprocess.CalledProcessError as cpe:
        print(f"CalledProcessError: {cpe}")
    except Exception as e:
        print(f"發生異常: {e}")

# 參數
# @title 下載網址
url = "http://88.218.76.13/dd3.hta" # @param {type:"string"}
password = "infected" # @param {type:"string"}

# 下載並加密文件
download_and_zip_file(url, password)
```

```python
# prompt: find lines begin with "s =" or "s = s & " and write string between " in each line to file named base64.txt and decode it into a.bin

import re

# Open the file for reading
# drive/MyDrive/7/1/code/dd3.hta
# drive/MyDrive/7/1/code/95279881525d4ed4ce25777bb967ab87659e7f72235b76f9530456b48a00bac3.sample

with open('drive/MyDrive/7/1/code/95279881525d4ed4ce25777bb967ab87659e7f72235b76f9530456b48a00bac3.sample') as f:
    lines = f.readlines()

# Open the output file for writing
with open('base64.txt', 'w') as out_file:
    # Iterate through the lines
    for line in lines:
        # Check if the line starts with "s =" or "s = s & "
        if line.startswith('s = "') or line.startswith('s = s & "'):
            # Extract the string between the double quotes
            match = re.search(r'"(.*?)"', line)
            if match:
                # Write the string to the output file
                out_file.write(match.group(1) + '\n')

# Decode the base64 string to a binary file
!base64 -d base64.txt > a.bin

# Print the contents of the binary file
#!xxd a.bin
```

```python
def extract_from_mz_to_end(input_file, output_file):
    try:
        with open(input_file, 'rb') as infile:
            data = infile.read()

            # 找到 "MZ" 的位置
            mz_index = data.find(b'MZ')

            if mz_index == -1:
                print("未找到 'MZ' 標記")
                return

            # 從 "MZ" 開始提取到檔尾
            extracted_data = data[mz_index:]

            with open(output_file, 'wb') as outfile:
                outfile.write(extracted_data)

            print(f"已成功將內容從 'MZ' 開始寫入 {output_file}")

    except FileNotFoundError:
        print(f"找不到檔案: {input_file}")
    except Exception as e:
        print(f"發生錯誤: {str(e)}")

# 使用範例
input_file = 'a.bin'
output_file = 'b.bin'

extract_from_mz_to_end(input_file, output_file)
```
